```python
# 0하고 10 사이를 100등분
x= np.linspace(0, 10, 100)
y= np.sin(x)
z= np.cos(x)

plt.plot(x,y,'--b')
plt.plot(x,z,'r')
# b는 파란색 --는 점선

plt.title('MP')
plt.xlabel('X')
```

```python
# figure : 큰 객체(최 상위 객체)
# axes :
```

```python
x= np.linspace(0, 10, 100)
y= np.sin(x)
z= np.cos(x)

plt.figure(figsize=(14,8))
plt.subplot(1,2,1)
plt.plot(x,y)

plt.subplot(1,2,2)
plt.plot(x,z)
plt.show()
```

```python
x= np.linspace(0, 10, 100)
y= np.sin(x)
z= np.cos(x)

plt.figure(figsize=(14,8))
plt.subplot(2,2,1)
plt.scatter(x,y)

plt.subplot(2,2,2)
plt.plot(x,z)

plt.subplot(2,2,3)
plt.bar(x,z)

plt.subplot(2,2,4)
plt.hist(x)
plt.show()
```

```python
import seaborn as sns

sns.__version__
```

```python
tips = sns.load_dataset("tips")
type(tips)
```

```python
tips.head()
```

```python
sns.scatterplot(data=tips, x= "total_bill",y="tip", hue="sex")
# 그래프 위에 표시되는 건 object
# hue로 성별 구분
```

```python
sns.barplot(data=tips, x= "sex", y='tip')
# 검정 선은 신뢰구간
```

```python
sns.countplot(data=tips, x= 'smoker')
```

```python
sns.histplot(data=tips, x= 'total_bill')
```

```python
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips)
```
